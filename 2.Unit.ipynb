```python
df_candidatures=pd.read_csv('Candidatures_final.csv')
df_candidatures.head()
```

```python
df_relacions=pd.read_csv('Relacions_final.csv')
df_relacions.head()
```

```python
from IPython import get_ipython
var_names = get_ipython().run_line_magic('who', '')
print(var_names)
```

```python
df=df_persones.merge(df_discapacitat, on="ID", how="left").merge(df_formatives, on="ID", how="left").merge(df_experiencia, on="ID", how="left").merge(df_seguiments, on="ID", how="left").merge(df_candidatures, on="ID", how="left").merge(df_relacions, on="ID", how="left")
df.head()
```

```python
df.duplicated(['ID']).sum()
```

```python
#eliminar duplicados de la columna ID
df.drop_duplicates(subset=['ID'], keep='first', inplace=True)
```

```python
df.duplicated(['ID']).sum()
```

```python
#ver porcentaje de nans por columna
df.isnull().sum()/df.shape[0]*100
#graficar porcentaje de nans
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
df.isnull().sum()/df.shape[0]*100
plt.bar(df.columns,df.isnull().sum()/df.shape[0]*100)
plt.xticks(rotation=45, ha='right')
plt.title('Porcentatje de valors faltants en les variables d''experiència laboral')
plt.show()
```

```python
#df a csv
df.to_csv('dades_totals.csv', index=False)
```

```python
# Seleccionar todas las filas donde la columna "Tipus de relacions actuals" contenga "Participant SAES"
filtro1 = df["Tipus de relacions actuals"].str.contains("Participant SAES")

# Seleccionar todas las filas donde la columna "Tipus de relació" sea igual a "Participant SAES"
filtro2 = df["Tipus de relació"] == "Participant SAES"

# Crear el nuevo DataFrame df_saes que incluye todas las filas que cumplen cualquiera de los dos filtros anteriores
df_saes = df[filtro1 | filtro2]

df_saes = df_saes.drop_duplicates()
```

```python
df_saes.shape
```

```python
#ver porcentaje de nans por columna
df_saes.isnull().sum()/df_saes.shape[0]*100
#graficar porcentaje de nans
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
df.isnull().sum()/df.shape[0]*100
plt.bar(df.columns,df.isnull().sum()/df.shape[0]*100)
plt.xticks(rotation=45, ha='right')
plt.title('Porcentatje de valors faltants en les variables d''experiència laboral')
plt.show()
```

```python
(df_saes.isnull().sum()/df_saes.shape[0]*100).sort_values(ascending=False)
```
